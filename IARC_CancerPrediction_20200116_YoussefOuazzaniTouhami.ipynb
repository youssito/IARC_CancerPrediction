```python
df_lung = pd.read_csv('lung_cancer.csv')
df_lung = df_lung.drop(columns = ['Unnamed: 26','Unnamed: 27'])
df_lung
```

```python
corr_matrix = df_lung.corr()
print(corr_matrix['Result'])
```

As we can see, Age is not sufficient and no other variable on its own is sufficient to explain the fact to have lung cancer.</br>

Using "Multi-class classification" a machine learning we'll try to predict if a person has cancer or not </br>
for this I chose 3 variables to explain using 4 variables : 
<ul>
    <li>Age</li>
    <li>Smoking</li>
    <li>Obesity</li>
    <li>Genetic Risk</li>
</ul>

```python
df_multiple_lung = df_lung[['Age','Smoking','Obesity','Genetic Risk','Result']]
df_multiple_lung
```

Let's split our dataset into training and testing data, i chose using k-nearest neighbors classifier because this model doesn't depend on the structure of the data

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split

X = df_multiple_lung[['Age','Smoking','Obesity','Genetic Risk']]
y = df_multiple_lung['Result']
  
# dividing X, y into train and test data (keep 60% for the train size and 40 for the test siez)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.4, random_state = 0)

#Let's train our model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 7).fit(X_train, y_train)

#Let's get our prediction
knn_predictions = knn.predict(X_test)
knn_predictions
```

```python
#Now we see how accurate was our model
accuracy = knn.score(X_test, y_test)
accuracy
```

Having an accuracy of 53%, we can say that these variables are not enough to predict lung cancer before diagnosys USING our model

```python
df_breast = pd.read_csv('breast_cancer.csv')
df_breast = df_breast.drop(columns = ['id'])
df_breast
```

M = malignant (cancerous) </br>
B = benign(non cancerous) </br>
We're gonna transform these data into 0 for non cancerous and 1 for cancerous

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
df_breast['diagnosis']=l.fit_transform(df_breast.diagnosis)
df_breast
```

We're going to use Support vector machine model

```python
x = df_breast.drop('diagnosis',axis=1)
y = df_breast['diagnosis']

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.svm import SVC
svm_model_linear = SVC(kernel = 'linear').fit(xtrain, ytrain)
svm_predictions = svm_model_linear.predict(xtest)

svm_predictions
```

```python
#Let's test the accuracy of our model
accuracy = svm_model_linear.score(xtest, ytest)
accuracy
```

Having an accuracy of 95%, we can conclude that breast cancer can be predicted using all the variables in our dataframe (without Age because it's not a good enough variable to predict breast cancer), but through a early diagnosis we can predict an find solutions to do so
